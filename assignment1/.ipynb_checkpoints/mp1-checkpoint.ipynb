```python
# On some implementations of matplotlib, you may need to change this value
IMAGE_SIZE = 72

def generate_a_drawing(figsize, U, V, noise=0.0):
    fig = plt.figure(figsize=(figsize,figsize))
    ax = plt.subplot(111)
    plt.axis('Off')
    ax.set_xlim(0,figsize)
    ax.set_ylim(0,figsize)
    ax.fill(U, V, "k")
    fig.canvas.draw()
    imdata = np.frombuffer(fig.canvas.tostring_rgb(), dtype=np.uint8)[::3].astype(np.float32)
    imdata = imdata + noise * np.random.random(imdata.size)
    plt.close(fig)
    return imdata

def generate_a_rectangle(noise=0.0, free_location=False, both=False):
    figsize = 1.0    
    U = np.zeros(4)
    V = np.zeros(4)
    if free_location:
        corners = np.random.random(4)
        top = max(corners[0], corners[1])
        bottom = min(corners[0], corners[1])
        left = min(corners[2], corners[3])
        right = max(corners[2], corners[3])
        
    else:
        side = (0.3 + 0.7 * np.random.random()) * figsize
        top = figsize/2 + side/2
        bottom = figsize/2 - side/2
        left = bottom
        right = top
    U[0] = U[1] = top
    U[2] = U[3] = bottom
    V[0] = V[3] = left
    V[1] = V[2] = right
    noisy = generate_a_drawing(figsize, U, V, noise)
    if both:
        without = generate_a_drawing(figsize, U, V, 0)
        return noisy, without
    return noisy


def generate_a_disk(noise=0.0, free_location=False, both=False):
    figsize = 1.0
    if free_location:
        center = np.random.random(2)
    else:
        center = (figsize/2, figsize/2)
    radius = (0.3 + 0.7 * np.random.random()) * figsize/2
    N = 50
    U = np.zeros(N)
    V = np.zeros(N)
    i = 0
    for t in np.linspace(0, 2*np.pi, N):
        U[i] = center[0] + np.cos(t) * radius
        V[i] = center[1] + np.sin(t) * radius
        i = i + 1
    noisy = generate_a_drawing(figsize, U, V, noise)
    if both:
        without = generate_a_drawing(figsize, U, V, 0)
        return noisy, without
    return noisy

def generate_a_triangle(noise=0.0, free_location=False, both=False):
    figsize = 1.0
    if free_location:
        U = np.random.random(3)
        V = np.random.random(3)
    else:
        size = (0.3 + 0.7 * np.random.random())*figsize/2
        middle = figsize/2
        U = (middle, middle+size, middle-size)
        V = (middle+size, middle-size, middle-size)
    noisy = generate_a_drawing(figsize, U, V, noise)
    if both:
        without = generate_a_drawing(figsize, U, V, 0)
        return noisy, without, [U[0], V[0], U[1], V[1], U[2], V[2]]
    return noisy, [U[0], V[0], U[1], V[1], U[2], V[2]]

```

```python
def generate_dataset_classification(nb_samples, noise=0.0, free_location=False):
    # Getting im_size:
    im_size = generate_a_rectangle().shape[0]
    X = np.zeros([nb_samples,im_size])
    Y = np.zeros(nb_samples)
    print('Creating data:')
    for i in range(nb_samples):
        category = np.random.randint(3)
        if category == 0:
            X[i] = generate_a_rectangle(noise, free_location)
        elif category == 1: 
            X[i] = generate_a_disk(noise, free_location)
        else:
            [X[i], V] = generate_a_triangle(noise, free_location)
        Y[i] = category
    print("Done!")
    X = (X + noise) / (255 + 2 * noise)
    return [X, Y]


def generate_test_set_classification():
    np.random.seed(42)
    [X_test, Y_test] = generate_dataset_classification(300, 20, True)
    Y_test = np_utils.to_categorical(Y_test, 3) 
    return [X_test, Y_test]

def generate_dataset_regression(nb_samples, noise=0.0):
    # Getting im_size:
    im_size = generate_a_triangle()[0].shape[0]
    X = np.zeros([nb_samples,im_size])
    Y = np.zeros([nb_samples, 6])
    print('Creating data:')
    for i in range(nb_samples):
        [X[i], Y[i]] = generate_a_triangle(noise, True)
    X = (X + noise) / (255 + 2 * noise)
    print("Done!")
    return [X, Y]

import matplotlib.patches as patches

def visualize_prediction(x, y):
    fig, ax = plt.subplots(figsize=(5, 5))
    I = x.reshape((IMAGE_SIZE,IMAGE_SIZE))
    ax.imshow(I, extent=[-0.15,1.15,-0.15,1.15],cmap='gray')
    ax.set_xlim([0,1])
    ax.set_ylim([0,1])

    xy = y.reshape(3,2)
    tri = patches.Polygon(xy, closed=True, fill = False, edgecolor = 'r', linewidth = 5, alpha = 0.5)
    ax.add_patch(tri)

    plt.show()

def generate_test_set_regression():
    np.random.seed(42)
    [X_test, Y_test] = generate_dataset_regression(300, 20)
    return [X_test, Y_test]


def generate_dataset_denoising(nb_samples, low = 20, high=40, free_location=False):
    # Getting im_size:
    im_size = generate_a_rectangle().shape[0]
    X_noise = np.zeros([nb_samples,im_size])
    X_without = np.zeros([nb_samples,im_size])
    print('Creating data:')
    random_noise = np.random.randint(low, high)
    for i in range(nb_samples):
        category = np.random.randint(3)
        if category == 0:
            X_noise[i], X_without[i] = generate_a_rectangle(random_noise, free_location, both=True)
        elif category == 1: 
            X_noise[i], X_without[i] = generate_a_disk(random_noise, free_location, both=True)
        else:
            X_noise[i], X_without[i], _ = generate_a_triangle(random_noise, free_location, both=True)
    print("Done!")
    return [X_noise/255, X_without/255]

labels = ['rectangle', 'circle', 'triangle']
```

### Simple Classification

#### Shape Identification : Constant Location

For this task, we're asked to try a fully connected neural network to classify the different shapes of the drawings : Rectangle, Circle or Triangle with a fixed localization. 

We firstly generate the dataset:

```python
[X_train, Y_train] = generate_dataset_classification(300, 20)
Y_train = Y_train.astype('int')
Y_train_one_hot = np.eye(3)[Y_train]
```

We define our model which is as we said a fully connected neural network with no hidden units, which is for our case, it's just a L2 regularized logistic regression on the flattened image.

```python
from keras import Sequential
from keras.layers import Dense, Conv2D, MaxPool2D, Flatten, Dropout
from keras.optimizers import sgd, Adam
from keras.losses import categorical_crossentropy
from keras import regularizers

def create_model():
    model = Sequential()
    model.add(Dense(units=3, input_shape=(72*72,), activation='softmax', kernel_regularizer=regularizers.l2(0.001)))
    return model
```

We know that the fully connected architecture is not well adapted to this kind of task but the purpose here is to test the SGD and Adam optimizers and visualize the weights learned to see the patterns learned for each shape.

**Simple Stochastic Gradient Descent**

We're using a simple sgd with a learning rate of 1e-3.

```python
model = create_model()
model.compile(optimizer = sgd(lr=0.001) , loss=categorical_crossentropy, metrics=['acc'])
model.fit(X_train, Y_train_one_hot, batch_size=32, epochs=40, verbose=0)
```

```python
scr_train = model.evaluate(X_train, Y_train_one_hot)
print('the loss on the train set is %.3f and the accuracy is %.2f'%(scr_train[0], 100*scr_train[1]))
```

Our model has no difficulty to fit well the train but get stuck in a local minima ( the loss function can't get under 0.35). 

Let's see if it works on a single circle image.

```python
X_test = generate_a_disk()
X_test = X_test.reshape(1, X_test.shape[0])
print('the predicted label is : ' + labels[np.argmax(model.predict(X_test))])
```

#### Visualization of the Solution

Let's visualize the weights it learned:

```python
weights_for_each = model.get_weights()[0]

fig = plt.figure(figsize=(15,6))

plt.subplot(1,3,1)
im = weights_for_each[:,0]
plt.title('weights for rectangles')
plt.imshow(im.reshape(IMAGE_SIZE,IMAGE_SIZE), cmap='gray')

plt.subplot(1,3,2)
im = weights_for_each[:,1]
plt.title('weights for circles')
plt.imshow(im.reshape(IMAGE_SIZE,IMAGE_SIZE), cmap='gray')

plt.subplot(1,3,3)
im = weights_for_each[:,2]
plt.title('weights for triangles')
plt.imshow(im.reshape(IMAGE_SIZE,IMAGE_SIZE), cmap='gray')

plt.show()
```

We can see that with a simple stochastic gradient descent, the weights learned don't follow any pattern which can mean that our model will have some hard time to generalize well.

**ADAM Optimizer**

Here, we're using an Adam optimizer with the same learning rate as before.

```python
model = create_model()
model.compile(optimizer = Adam(lr=0.001) , loss=categorical_crossentropy, metrics=['acc'])
model.fit(X_train, Y_train_one_hot, batch_size=32, epochs=40, verbose=0)
scr_train = model.evaluate(X_train, Y_train_one_hot)
print('the loss on the train set is %.3f and the accuracy is %.2f'%(scr_train[0], 100*scr_train[1]))
```

The adam optimizer reaches a better minima and even get us a 100% accuracy in the train.

```python
X_test = generate_a_disk()
X_test = X_test.reshape(1, X_test.shape[0])
print('the predicted label is : ' + labels[np.argmax(model.predict(X_test))])
```

#### Visualization of the Solution

```python
weights_for_each = model.get_weights()[0]

fig = plt.figure(figsize=(15,6))

plt.subplot(1,3,1)
im = weights_for_each[:,0]
plt.title('weights for rectangles')
plt.imshow(im.reshape(IMAGE_SIZE,IMAGE_SIZE), cmap='gray')

plt.subplot(1,3,2)
im = weights_for_each[:,1]
plt.title('weights for circles')
plt.imshow(im.reshape(IMAGE_SIZE,IMAGE_SIZE), cmap='gray')

plt.subplot(1,3,3)
im = weights_for_each[:,2]
plt.title('weights for triangles')
plt.imshow(im.reshape(IMAGE_SIZE,IMAGE_SIZE), cmap='gray')

plt.show()
```

We can see that the patterns learned with the Adam optimizer are meaningful and can help our model generalize better.

### A more difficult classification problem

#### Shape Identification : Free Location

The drawings are no longed centered in the image which make this task slightly more diffuclt, we switched to Convolutional Neural Networks which are clearly more adapted to image classification tasks.

```python
[X_train, Y_train] = generate_dataset_classification(300, 20, True)
X_train = X_train.reshape((-1, 72,72,1))
Y_train = Y_train.astype('int')
Y_train_one_hot = np.eye(3)[Y_train]
```

```python
[X_test, Y_test] = generate_test_set_classification()
X_test = X_test.reshape((-1, 72,72,1))
```

```python
def create_cnn_classification():
    model = Sequential()
    model.add(Conv2D(8, (3,3), activation='relu', input_shape=(72,72,1)))
    model.add(MaxPool2D())
    model.add(Conv2D(16, (3,3), activation='relu', kernel_regularizer=regularizers.l2(0.01)))
    model.add(MaxPool2D())
    model.add(Conv2D(16, (3,3), activation='relu'))
    model.add(MaxPool2D())
    model.add(Flatten())
    model.add(Dense(3, activation='softmax', kernel_regularizer=regularizers.l2(0.01)))
    return model
```

The architecture of the model can be seen bellow:

```python
model = create_cnn_classification()
model.summary()
```

```python
model.compile(optimizer = Adam(lr=0.001) , loss=categorical_crossentropy, metrics=['accuracy'])
model.fit(X_train, Y_train_one_hot, batch_size=32, epochs=60, verbose=0)
scr_train = model.evaluate(X_train, Y_train_one_hot)
print('the loss on the test set is %.3f and the accuracy is %.2f'%(scr_train[0], 100*scr_train[1]))
scr_test = model.evaluate(X_test, Y_test)
print('the loss on the train set is %.3f and the accuracy is %.2f'%(scr_test[0], 100*scr_test[1]))
```

We can see that even for this slightly difficult task, we could build a model that did pretty well on the test data.

### Regression Problem

#### Triangle Vertices Identification

The purpose of this task is, given a triangle, predict the plane position (x,y) of the 3 vertices. 

It can be formulated as a regression problem to predict 6 real numbers between 0 and 1. The use of a sigmoid function can be though of but the saturation capacity of the sigmoid may lead to poor results, so we're gonna stick with value clipping.

When we investigate how the predefined utilities return the vertices positions, we can see that these 3 positions are given randomly and if we want to learn on these points, our model will learn on noise because of the random permutations of the given positions. 

To solve this problem, we need to sort these vertices ( we sorted them by their respective abscisses ) to make the task of learning easier for our model. 

```python
def vertice_sorting(triangle):
    triangle += -0.5
    v1, v2, v3 = triangle[:2], triangle[2:4], triangle[4:]
    vs = sorted([list(v1), list(v2), list(v3)])
    return np.array(vs).flatten()
```

We create a bigger training dataset (1000 samples) and center the values on 0.0 by deducting 0.5 from the inputs and the labels to get a well posed problem ( easier to optimize ).

```python
[X_train, Y_train] = generate_dataset_regression(1000, 20)
X_train = 0.5 - X_train
X_train = X_train.reshape((-1, 72,72,1))
Y_train = np.apply_along_axis(vertice_sorting, axis=1, arr= Y_train)
```

```python
[X_test, Y_test] = generate_test_set_regression()
X_test = 0.5 - X_test
X_test = X_test.reshape((-1, 72,72,1))
Y_test = np.apply_along_axis(vertice_sorting, axis=1, arr= Y_test)
```

```python
visualize_prediction(X_train[0], Y_train[0]+0.5)
```

```python
def create_cnn_regression():
    model = Sequential()
    model.add(Conv2D(8, (3,3), activation='relu', input_shape=(72,72,1)))
    model.add(MaxPool2D())
    model.add(Conv2D(16, (3,3), activation='relu'))
    model.add(MaxPool2D())
    model.add(Conv2D(16, (3,3), activation='relu'))
    model.add(MaxPool2D())
    model.add(Flatten())
    model.add(Dense(6))
    return model
```

```python
cnn = create_cnn_regression()
cnn.compile(optimizer = Adam(lr=0.001) , loss='mean_absolute_error')
cnn.summary()
```

```python
cnn.fit(X_train, Y_train, batch_size=32, epochs=60, verbose=0)

scr_train = cnn.evaluate(X_train, Y_train)
print('the loss on the test set is %.3f'%(scr_train))
scr_test = cnn.evaluate(X_test, Y_test)
print('the loss on the train set is %.3f'%(scr_test))
```

Our model fit well the training data, let's visualize some results.

```python
Y_pred_test = cnn.predict(X_test)
Y_pred_train = cnn.predict(X_train)
```

```python
seed(2)
idx = np.random.randint(0, high=300)
visualize_prediction(X_train[idx], np.clip(Y_pred_train[idx] +0.5, 0.0, 1.0))
```

We can see that we get some pretty descent results.

### Image Denoising

For this particular task, we define an AutoEncoder with a Hourglass architecture described below:

```python
from keras.layers import Conv2DTranspose, UpSampling2D
def create_denoiser():
    model = Sequential()
    model.add(Conv2D(8, (3,3), activation='relu', input_shape=(72,72,1), padding='same'))
    model.add(MaxPool2D())
    model.add(Conv2D(8, (3,3), activation='relu', padding='same'))
    model.add(MaxPool2D())
    model.add(Conv2D(8, (3,3), activation='relu', padding='same'))
    model.add(MaxPool2D())
    model.add(Conv2DTranspose(8, (3,3), activation='relu', padding='same'))
    model.add(UpSampling2D())
    model.add(Conv2DTranspose(8, (3,3), activation='relu', padding='same'))
    model.add(UpSampling2D())
    model.add(Conv2DTranspose(1, (3,3), activation='relu', padding='same'))
    model.add(UpSampling2D())
    model.add(Conv2DTranspose(1, (3,3), activation='sigmoid', padding='same'))
    return model
```

We create a randomly noised dataset from 20 to a 60 noise amplitude. We also take **1.0 - X_train** as it makes the optimization problem easier.

```python
X_noise, X_without = generate_dataset_denoising(600, high=60)
X_noise, X_without = 1.0 - X_noise.reshape(-1,72,72,1), 1.0 - X_without.reshape(-1,72,72,1)
```

```python
plt.figure(figsize=(12,5))
idx = np.random.randint(0, high=600)
im, noisy = X_without[idx], X_noise[idx]
plt.subplot(1,2,1)
plt.imshow(noisy.reshape(IMAGE_SIZE,IMAGE_SIZE), cmap='gray')
plt.subplot(1,2,2)
plt.imshow(im.reshape(IMAGE_SIZE,IMAGE_SIZE), cmap='gray')
plt.show()
```

Let's define our train/test split :

```python
split = 550
train_noise, train_true = X_noise[:split], X_without[:split]
test_noise, test_true = X_noise[split:], X_without[split:]
```

We create our model :

```python
denoiser = create_denoiser()
denoiser.compile(optimizer = Adam(lr=0.01) , loss='binary_crossentropy')
denoiser.summary()
```

```python
denoiser.fit(train_noise, train_true, batch_size=32, epochs=40, verbose=0)
scr_train = denoiser.evaluate(train_noise, train_true)
print('the loss on the train set is %.3f'%(scr_train))
scr_test = denoiser.evaluate(test_noise, test_true)
print('the loss on the test set is %.3f'%(scr_test))
```

We can see that we have a pretty good denoising results.

```python
idx = np.random.randint(50)

fig = plt.figure(figsize=(15,6))

plt.subplot(1,3,1)
plt.title('Noisy Image')
plt.imshow(1.0 - test_noise[[idx]].reshape(IMAGE_SIZE,IMAGE_SIZE), cmap='gray')

plt.subplot(1,3,2)
plt.title('Ground Truth')
plt.imshow(1.0 - test_true[[idx]].reshape(IMAGE_SIZE,IMAGE_SIZE), cmap='gray')

plt.subplot(1,3,3)
plt.title('Denoised Image')
plt.imshow(1.0 - denoiser.predict(test_noise[[idx]]).reshape(IMAGE_SIZE,IMAGE_SIZE), cmap='gray')

plt.show()
```
